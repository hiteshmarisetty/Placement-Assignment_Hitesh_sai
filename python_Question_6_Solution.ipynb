# Insights to be drawn -

- Get all Pokemons whose spawn rate is less than 5%
- Get all Pokemons that have less than 4 weaknesses
- Get all Pokemons that have no multipliers at all
- Get all Pokemons that do not have more than 2 evolutions
- Get all Pokemons whose spawn time is less than 300 seconds.

**Note:**
spawn time format is "05:32”, so assume “minute: second” format and perform the analysis.
- Get all Pokemon who have more than two types of capabilities

## 1. Get all Pokemons whose spawn rate is less than 5%

```python
ans_1 = df[df['spawn_chance'] < 0.05][['name','spawn_chance']]
ans_1.head()
```

```python
plt.figure(figsize=(15, 6))
plt.bar(ans_1['name'], ans_1['spawn_chance'])
plt.xlabel('Pokemon')
plt.ylabel('Spawn Chance (%)')
plt.title('Pokemon with Spawn Rate Less than 5%')
plt.xticks(rotation=90)
plt.show()
```

```python
df['weaknesses'].head()   
```

# 2. Get all Pokemons that have less than 4 weaknesses

```python
ans_2 = df[df['weaknesses'].str.count(',') < 3][['name','weaknesses']]
ans_2.head()
```

```python
plt.figure(figsize=(20, 6))
plt.bar(ans_2['name'], ans_2['weaknesses'])
plt.xlabel('Pokemon')
plt.ylabel('Weaknesses (%)')
plt.title('Pokemons that have less than 4 weaknesses')
plt.xticks(rotation=90)
plt.show()
```

# 3. Get all Pokemons that have no multipliers at all

```python
df['multipliers'].head()
```

```python
ans_3 = df[df['multipliers'].isnull()][['name','multipliers']]
ans_3.head()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(ans_3['name'], ans_3['multipliers'])
plt.xlabel('Pokemon')
plt.ylabel('Multipliers (%)')
plt.title('Pokemons that have no multipliers at all')
plt.xticks(rotation=90)
plt.show()
```

# 4. Get all Pokemons that do not have more than 2 evolutions

```python
ans_4 = df[df['next_evolution'].apply(lambda x: len(x) if isinstance(x, list) else 0) <= 2][['name','next_evolution']].head()
ans_4
```

# 5. Get all Pokemons whose spawn time is less than 300 seconds.

```python
# 300 seconds means ===> 06:00 minute:second
pd.to_datetime(df['spawn_time'].apply(lambda x: x.strftime('%H:%M:%S') if pd.notnull(x) else ''))
ans_5 = df[df['spawn_time'].apply(lambda x: x.minute < 6 or (x.minute == 6 and x.second == 0))][['name','spawn_time']].head()
ans_5
```

# 6. Get all Pokemon who have more than two types of capabilities

```python
ans_6 = df[df['type'].str.count(',') <= 2][['name','type']]
ans_6.head()
```

```python
plt.figure(figsize=(25, 6))
plt.bar(ans_6['name'], ans_6['type'])
plt.xlabel('Pokemon')
plt.ylabel('Type (%)')
plt.title('Pokemon who have more than two types of capabilities')
plt.xticks(rotation=90)
plt.show()
```
