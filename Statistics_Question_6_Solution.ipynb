## Question_6_Solution:

**Q-6.** A pharmaceutical company is developing a drug that is supposed to reduce blood
pressure. They conduct a clinical trial with 100 patients and record their blood
pressure before and after taking the drug. The company wants to know if the change
in blood pressure follows a normal distribution.

https://drive.google.com/file/d/1mCjtYHiX--mMUjicuaP2gH3k-SnFxt8Y/view?usp=share_

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_csv('Q3.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
before_bp = df[' Blood Pressure Before (mmHg)']
after_bp = df[' Blood Pressure After (mmHg)']
```

```python
# Generate Q-Q plot for blood pressure before
sm.qqplot(before_bp, line='s')
plt.title('Q-Q Plot - Blood Pressure Before')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.show()

# Generate Q-Q plot for blood pressure after
sm.qqplot(after_bp, line='s')
plt.title('Q-Q Plot - Blood Pressure After')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.show()
```

Observation: 

By Observaing both the Q-Q plot, It can be concluded that After testing new drug on BP patients. There is slight improvement in Patients Blood-Pressure. So, Stricty Yes, It is following somewhat normal distribution.
